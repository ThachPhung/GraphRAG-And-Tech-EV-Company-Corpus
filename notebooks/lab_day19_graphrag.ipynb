Notebook demo các bước: build index, xem graph, hỏi đáp Flat RAG vs GraphRAG.

```python
import sys
from pathlib import Path

ROOT = Path('..').resolve()
sys.path.append(str(ROOT / 'src'))
ROOT
```

```python
from data_loader import load_documents

docs = load_documents(ROOT / 'dataset')
len(docs), docs[0].title
```

```python
# Chạy cell này nếu chưa build index từ terminal.
from build_index import build

build(str(ROOT / 'dataset'), str(ROOT / 'outputs'), use_openai=False)
```

```python
from flat_rag import FlatRetriever
from graph_rag import load_graph, answer_with_graph

retriever = FlatRetriever.load(ROOT / 'outputs' / 'flat_retriever.pkl')
graph = load_graph(ROOT / 'outputs' / 'knowledge_graph.json')
graph.number_of_nodes(), graph.number_of_edges()
```

```python
question = 'Why did US EV sales growth slow in Q1 2024 and how was Tesla involved?'

print('Flat RAG top docs:')
for result in retriever.search(question, top_k=3):
    print(result.doc_id, round(result.score, 3), result.title)

answer = answer_with_graph(graph, docs, question, retriever=retriever)
print('\nGraphRAG answer:')
print(answer.answer)
print('\nGraph paths:')
for path in answer.paths[:5]:
    print('-', path)
```
